# Deskripsi Kasus

Merupakan kumpulan data Kesehatan Tidur dan Gaya Hidup terdiri dari 400 baris dan 13 kolom, yang mencakup berbagai variabel terkait tidur dan kebiasaan sehari-hari. Data tersebut digunakan untuk mengetahui faktor-faktor apa saja yang mempengaruhi pola tidur berdasarkan variabel - variabel yang akan dianalisis, yakni sebagai berikut:

1. Person ID : pengenal untuk setiap individu (1,2,3,4,...)

2. Gender (Male or Female)

3. Age (Continuous)

4. Occupation (Software Engineer, Doctor, Sales Representative, Teacher, Nurse, Engineer, Accountant, Scientist, Lawyer,   Salesperson, and Manager)

5. Sleep Duration (jam) : jumlah jam orang tidur per hari.

6. Quality of Sleep (skala: 1-10)

7. Physical Activity Level (menit/hari): Jumlah menit orang tersebut melakukan aktivitas fisik setiap hari.

8. Stress Level (skala: 1-10)

9. BMI Category (Underweight, Normal, Overweight, Obese)

10. Blood Pressure (sistolik/diastolik) : misalnya 126/83, 125/80, 140/90, ...

11. Heart Rate (bpm): Denyut jantung istirahat orang tersebut dalam denyut per menit( 75, 80, 77, 79,..)

12. Daily Steps: Jumlah langkah yang diambil orang tersebut per hari (10000, 4200, 3500,..)

13. Sleep Disorder: Ada atau tidaknya gangguan tidur pada seseorang (Tidak Ada, Insomnia, Sleep Apnea).

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 8))
sns.scatterplot(
    x='Age', 
    y='Sleep Duration', 
    size='Quality of Sleep', 
    sizes=(20, 200), 
    hue='Gender', 
    palette='coolwarm', 
    data=df, 
    legend='full', 
    alpha=0.6
)
plt.title('Diagram Gelembung: Umur vs. Durasi Tidur dengan Kualitas Tidur sebagai Ukuran Gelembung')
plt.xlabel('Umur')
plt.ylabel('Durasi Tidur (menit)')
plt.legend(title='Jenis Kelamin', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
#boxplot
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='Gender', y='Quality of Sleep', data=df, palette='pastel')
plt.title('Kualitas Tidur Berdasarkan Jenis Kelamin')
plt.xlabel('Jenis Kelamin')
plt.ylabel('Kualitas Tidur')
plt.show()
```

```python
#barchart
```

```python
# Menghitung jumlah setiap kategori BMI
bmi_counts = df['BMI Category'].value_counts()
plt.figure(figsize=(12, 6))
sns.barplot(x=bmi_counts.values, y=bmi_counts.index, palette='viridis')
plt.title('Jumlah Orang Berdasarkan Kategori BMI')
plt.xlabel('Jumlah')
plt.ylabel('Kategori BMI')
plt.show()
```

```python
#4. Line Chart: Melihat tren data time series
#Misalnya, kita buat line chart berdasarkan umur untuk melihat tren
plt.figure(figsize=(12, 6))
df_sorted = df.sort_values('Age')
sns.lineplot(x='Age', y='Sleep Duration', data=df_sorted, marker='o', color='coral')
plt.title('Tren Durasi Tidur Berdasarkan Umur')
plt.xlabel('Umur')
plt.ylabel('Durasi Tidur (menit)')
plt.show()
```

```python
# 5. Scatter Plot: Melihat hubungan antara dua variabel numerik
plt.figure(figsize=(12, 6))
sns.scatterplot(x='Physical Activity Level', y='Sleep Duration', data=df, hue='Gender', palette='coolwarm')
plt.title('Hubungan antara Tingkat Aktivitas Fisik dan Durasi Tidur')
plt.xlabel('Tingkat Aktivitas Fisik')
plt.ylabel('Durasi Tidur (menit)')
plt.legend(title='Jenis Kelamin')
plt.show()
```
