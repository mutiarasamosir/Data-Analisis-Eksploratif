# Deskripsi Kasus

Merupakan kumpulan data Kesehatan Tidur dan Gaya Hidup terdiri dari 374 baris dan 13 kolom, yang mencakup berbagai variabel terkait tidur dan kebiasaan sehari-hari. Data tersebut digunakan untuk mengetahui faktor-faktor apa saja yang mempengaruhi pola tidur berdasarkan variabel - variabel yang akan dianalisis, yakni sebagai berikut:

1. Person ID : pengenal untuk setiap individu (1,2,3,4,...)

2. Gender (Male or Female)

3. Age (Continuous)

4. Occupation (Software Engineer, Doctor, Sales Representative, Teacher, Nurse, Engineer, Accountant, Scientist, Lawyer,   Salesperson, and Manager)

5. Sleep Duration (jam) : jumlah jam orang tidur per hari.

6. Quality of Sleep (skala: 1-10)

7. Physical Activity Level (menit/hari): Jumlah menit orang tersebut melakukan aktivitas fisik setiap hari.

8. Stress Level (skala: 1-10)

9. BMI Category (Underweight, Normal, Overweight, Obese)

10. Blood Pressure (sistolik/diastolik) : misalnya 126/83, 125/80, 140/90, ...

11. Heart Rate (bpm): Denyut jantung istirahat orang tersebut dalam denyut per menit( 75, 80, 77, 79,..)

12. Daily Steps: Jumlah langkah yang diambil orang tersebut per hari (10000, 4200, 3500,..)

13. Sleep Disorder: Ada atau tidaknya gangguan tidur pada seseorang (Tidak Ada, Insomnia, Sleep Apnea).

```python
import pandas as pd

df = pd.read_csv(r"C:\\Users\\Documents\\semester 3\\data analisis eksploratif\\dataset1.csv", sep=';')

df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd



# Visualisasi 1: Gender
plt.figure(figsize=(4, 3))
sns.countplot(x='Gender', data=df, palette='Set2')
plt.title('Distribution of Gender')
plt.show()

# Visualisasi 2: Age
plt.figure(figsize=(6, 4))
sns.histplot(df['Age'], bins=10, color='skyblue')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# Visualisasi 3: Sleep Duration
plt.figure(figsize=(4, 3))
sns.boxplot(x='Sleep Duration', data=df, color='lightgreen')
plt.title('Sleep Duration Distribution')
plt.show()

# Visualisasi 4: Quality of Sleep
plt.figure(figsize=(6, 4))
sns.histplot(df['Quality of Sleep'], bins=10, color='lightcoral')
plt.title('Quality of Sleep Distribution')
plt.xlabel('Quality of Sleep (1-10)')
plt.ylabel('Frequency')
plt.show()

# Visualisasi 5: Physical Activity Level
plt.figure(figsize=(6, 4))
sns.histplot(df['Physical Activity Level'], bins=10, color='lightblue')
plt.title('Physical Activity Level Distribution')
plt.xlabel('Physical Activity Level (minutes/day)')
plt.ylabel('Frequency')
plt.show()

# Visualisasi 6: Stress Level
plt.figure(figsize=(6, 4))
sns.histplot(df['Stress Level'], bins=10, color='lightgray')
plt.title('Stress Level Distribution')
plt.xlabel('Stress Level (1-10)')
plt.ylabel('Frequency')
plt.show()

# Visualisasi 7: BMI Category
plt.figure(figsize=(6, 4))
sns.countplot(x='BMI Category', data=df, palette='muted')
plt.title('BMI Category Distribution')
plt.xlabel('BMI Category')
plt.ylabel('Count')
plt.show()

# Visualisasi 8: Blood Pressure
df[['Systolic', 'Diastolic']] = df['Blood Pressure'].str.split('/', expand=True)
df['Systolic'] = df['Systolic'].astype(int)
df['Diastolic'] = df['Diastolic'].astype(int)

plt.figure(figsize=(4, 3))
sns.scatterplot(x='Systolic', y='Diastolic', data=df, color='orange')
plt.title('Scatter Plot of Blood Pressure (Systolic vs Diastolic)')
plt.show()

# Visualisasi 9: Heart Rate
plt.figure(figsize=(6, 4))
sns.histplot(df['Heart Rate'], bins=10, color='purple')
plt.title('Heart Rate Distribution')
plt.xlabel('Heart Rate (bpm)')
plt.ylabel('Frequency')
plt.show()

# Visualisasi 10: Daily Steps
plt.figure(figsize=(6, 4))
sns.histplot(df['Daily Steps'], bins=10, color='salmon')
plt.title('Daily Steps Distribution')
plt.xlabel('Daily Steps')
plt.ylabel('Frequency')
plt.show()

# Visualisasi 11: Sleep Disorder 
plt.figure(figsize=(4, 3))
sns.countplot(x='Sleep Disorder', data=df, palette='Set1')
plt.title('Distribution of Sleep Disorders')
plt.show()
```

```python
disorder_by_gender = df.groupby(['Gender', 'Sleep Disorder']).size().unstack()
disorder_percentage = disorder_by_gender.div(disorder_by_gender.sum(axis=1), axis=0) * 100
ax = disorder_by_gender.plot(kind='bar', stacked=True, figsize=(10, 6), rot=0) 
plt.title('Perbandingan Sleep Disorder Berdasarkan Gender')
plt.xlabel('Gender')
plt.ylabel('Jumlah Orang')
plt.show()
print(disorder_percentage)
```

```python
# Scatter Plot untuk hubungan antara Stress Level dan Quality of Sleep
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Stress Level', y='Quality of Sleep', data=df, color='teal')
plt.title('Relationship between Stress Level and Quality of Sleep')
plt.xlabel('Stress Level (1-10)')
plt.ylabel('Quality of Sleep (1-10)')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 8))
sns.scatterplot(
    x='Age', 
    y='Sleep Duration', 
    size='Quality of Sleep', 
    sizes=(20, 200), 
    hue='Gender', 
    palette='coolwarm', 
    data=df, 
    legend='full', 
    alpha=0.6
)
plt.title('Diagram Gelembung: Umur vs. Durasi Tidur dengan Kualitas Tidur sebagai Ukuran Gelembung')
plt.xlabel('Umur')
plt.ylabel('Durasi Tidur (menit)')
plt.legend(title='Jenis Kelamin', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# Menghitung jumlah setiap kategori BMI
bmi_counts = df['BMI Category'].value_counts()
plt.figure(figsize=(12, 6))
sns.barplot(x=bmi_counts.values, y=bmi_counts.index, palette='viridis')
plt.title('Jumlah Orang Berdasarkan Kategori BMI')
plt.xlabel('Jumlah')
plt.ylabel('Kategori BMI')
plt.show()
```

```python
#4. Line Chart: Melihat tren data time series
#Misalnya, kita buat line chart berdasarkan umur untuk melihat tren
plt.figure(figsize=(12, 6))
df_sorted = df.sort_values('Age')
sns.lineplot(x='Age', y='Sleep Duration', data=df_sorted, marker='o', color='coral')
plt.title('Tren Durasi Tidur Berdasarkan Umur')
plt.xlabel('Umur')
plt.ylabel('Durasi Tidur (menit)')
plt.show()
```

```python
# 5. Scatter Plot: Melihat hubungan antara dua variabel numerik
plt.figure(figsize=(12, 6))
sns.scatterplot(x='Physical Activity Level', y='Sleep Duration', data=df, hue='Gender', palette='coolwarm')
plt.title('Hubungan antara Tingkat Aktivitas Fisik dan Durasi Tidur')
plt.xlabel('Tingkat Aktivitas Fisik')
plt.ylabel('Durasi Tidur (menit)')
plt.legend(title='Jenis Kelamin')
plt.show()
```
